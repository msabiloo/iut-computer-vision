# آستانه‌گذاری (Thresholding)

برای آستانه‌گذاری، در ساده‌ترین حالت یک مقدار آستانه T انتخاب می‌کنیم

برای هر پیکسل از تصویر خاکستری، اگر مقدار پیکسل برابر یا بزرگ‌تر از آستانه بود، مقدار آن پیکسل را ۱ (255) .قرار می‌دهیم، در غیر این‌صورت ۰ قرار می‌دهیم

برای مثال برای تصویر زیر، مقدار آستانه را برابر با ۹۷ قرار می‌دهیم

```python
# import necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# to show image in notebook
%matplotlib inline
```

```python
img = cv2.imread('images/Picture1.jpg')

plt.axis('off')
plt.imshow(img, cmap='gray')
```

ابتدا تصویر رنگی را به تصویر خاکستری تبدیل می‌کنیم

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(img_gray, cmap='gray')
```

یک ماتریس به اندازه تصویر با مقادیر ۰ ایجاد می‌کنیم

```python
T = 97

height, width = img_gray.shape

img_binary = np.zeros((height, width))
```

در یک حلقه تو در تو مقدار هر پیکسل را با آستانه مقایسه می‌کنیم

```python
for i in range(height):
  for j in range(width):
    if img_gray[i, j] >= T:
      img_binary[i, j] = 255
    else:
      img_binary[i, j] = 0
```

```python
plt.axis('off')
plt.imshow(img_binary, cmap='gray')
```

در OpenCV با استفاده از دستور threshold این کار را می توانیم انجام دهیم.
این دستور در خورجی ۲ مقدار را باز می‌گرداند:
* مقدار اول برابر با آستانه (T) است
* مقدار دوم تصویر باینری

```python
_, img_binary = cv2.threshold(img_gray, T, 255, cv2.THRESH_BINARY)

plt.axis('off')
plt.imshow(img_binary, cmap='gray')
```
