# تحلیل دونمایی هیستوگرام (Bimodal Histogram Analysis)



با تحلیل هیستوگرام یک تصویر می‌توان آستانه را بدست آورد.

اگر فرض کنیم پس‌زمینه تصویر حول یک مقدار خاص و پیش‌زمینه تصویر نیز حول یک مقدار خاص در نمودار هیستوگرام باشد، به عبارت دیگر هیستوگرام تصویر ۲ پیک اصلی یا ۲ قله داشته باشد، آنگاه کمترین مقدار بین ۲ قله مقدار آستانه خواهد بود.

به کمترین مقدار بین دو قله (یا ۲ پیک اصلی) نقطه آنتی مد (Antimode) می‌گویند

برای مثال می‌خواهیم آستانه تصویر زیر را از روش تحلیل دونمایی هیستوگرام بدست آوریم

```python
# import necessary libraries
import cv2
import matplotlib.pyplot as plt

# to show image in notebook
%matplotlib inline
```

```python
img = cv2.imread('images/Picture1_XXV.jpg')

plt.axis('off')
plt.imshow(img, cmap='gray')
```

ابتدا تصویر رنگی را به تصویر خاکستری تبدیل می‌کنیم

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(img_gray, cmap='gray')
```

نمودار هیستوگرام آن را رسم می‌کنیم

```python
histogram = cv2.calcHist([img], [0], None, [256], [0, 256])

plt.plot(histogram)
plt.xlim([0, 256])
plt.ylim([0, 110])
plt.show()
```

همانگونه که در نمودار هیستوگرام مشخص است ۱۰۰ کمترین مقدار را میان دو قله دارد؛ بنابراین آستانه تصویر را می‌توانیم ۱۰۰ انتخاب کنیم

```python
# Threshold
T = 100

_, img_binary = cv2.threshold(img_gray, T, 255, cv2.THRESH_BINARY)

plt.axis('off')
plt.imshow(img_binary, cmap='gray')
```
