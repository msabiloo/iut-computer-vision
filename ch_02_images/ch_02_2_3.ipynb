# تصویرهای رنگی (Color Images)
تصویرهای خاکستری یک کانال داده دارند ولی تصویرهای رنگی می‌توانند چندین کانال داده داشته باشند

## تصویرهای RGB
در تصویرهای RGB به جای استفاده از یک کانال داده از سه کانال داده استفاده می‌شود

در OpenCV اما ترتیب کانال‌های RGB متفاوت است.

ترتیب کانال‌های رنگی در OpenCV:
1. آبی (Blue)
2. سبز (Green)
3. قرمز (Red)

با استفاده از متد split می‌توانیم کانال‌های داده تصاویر RGB را از هم جدا کنیم:

```python
import cv2 as cv
from matplotlib import pyplot as plt
import math

img = cv.imread('images/coat_hanger.jpg')

(b_channel, g_channel, r_channel) = cv.split(img)

# plt.axis('off')
# plt.imshow(b_channel, cmap='gray')
# plt.show()

f, axarr = plt.subplots(2, 2)

axarr[0, 0].imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
axarr[0, 0].set_title('RGB Image')
axarr[0, 0].axis('off')

axarr[0, 1].imshow(r_channel, cmap='gray')
axarr[0, 1].set_title('Red Channel')
axarr[0, 1].axis('off')

axarr[1, 0].imshow(g_channel, cmap='gray')
axarr[1, 0].set_title('Green Channel')
axarr[1, 0].axis('off')

axarr[1, 1].imshow(b_channel, cmap='gray')
axarr[1, 1].set_title('Blue Channel')
axarr[1, 1].axis('off')

f.subplots_adjust(hspace=0.3)
plt.show()
```

به روش دیگری نیز می‌توانیم کانال‌های داده تصاویر RGB را از هم جدا کنیم:

```python
plt.axis('off')
plt.imshow(img[:,:,0], cmap='gray')
plt.show()
```

برای تبدیل تصویر RGB به تصویر خاکستری، می‌توان از تابع cvtColor استفاده کرد:

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(gray_img, cmap='gray')
plt.show()
```

روش دیگر تبدیل تصویر RGB به تصویر خاکستری، استفاده از فرمول زیر است:

Y = مقدار هر پیکسل در تصویر خاکستری

`Y = (0.299 * R) + (0.587 * G) + (0.114 * B)`

مثلا اگر بخواهیم اولین پیکسل از تصویر اصلی را به خاکستری تبدیل کنیم:

```python
Y = math.floor((0.299 * img[0, 0, 2]) + (0.587 * img[0, 0, 1]) + (0.114 * img[0, 0, 0]))

print(Y)
```

به طور مشابه همین کار را برای تمام پیکسل‌ها انجام می‌دهیم:

```python
gray_img = img[:,:,0]

for i in range(0, gray_img.shape[0]):
    for j in range(0, gray_img.shape[1]):
        gray_img[i, j] = math.floor((0.299 * img[i, j, 2]) + (0.587 * img[i, j, 1]) + (0.114 * img[i, j, 0]))

plt.axis('off')
plt.imshow(gray_img, cmap='gray')
plt.show()
```
