# کمی‌سازی (Quantisation)
به زبان ساده یعنی هر پیکسل از تصویر از مقدار ۰ تا چه مقداری می تواند داشته باشد.

برای مثال زمانی که در مورد یک تصویر خاکستری ۸ بیتی صحبت می کنیم، یعنی هر پیکسل از تصویر عددی بین ۰ تا ۲۵۵ می‌تواند داشته باشد.

می خواهیم یک تصویر رنگی را ابتدا به یک تصویر خاکستری تبدیل کنیم و سپس مقدار عددی که هر پیکسل می تواند بگیرد، عددی بین ۰ و ۶۳ باشد.

ابتدا تصویر را فراخوانی می کنیم و سپس آن را به یک تصویر خاکستری تبدیل می‌کنیم:

```python
import cv2 as cv
from matplotlib import pyplot as plt
import math

img = cv.imread('images/church.jpg')

gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(gray_img, cmap='gray')
plt.show()
```

برای اینکه ببینیم تصویر تا چه مقداری کوانتیزه شده است، از پاپرتی dtype استفاده می‌کنیم:

```python
print(gray_img.dtype)
```

uint8 به معنی این است که هر پیکسل یک عدد صحیح مثبت از ۰ تا ۲۵۵ است.

یکی از ساده‌ترین راه‌ها برای تبدیل مقدار هر پیکسل از عددی بین ۰ تا ۲۵۵ به عددی بین ۰ تا ۶۳، به روش زیر است:

۲ به توان ۸ را بر ۲ به توان ۶ تقسیم می‌کنیم. حاصل این تقسیم عدد ۴ است.

 مقدار هر پیکسل از تصویر اصلی را بر عد ۴ تقسیم می‌کنیم و جز صحیح آن را به عنوان مقدار جدید پیکسل در نظر می‌گیریم

برای اولین پیکسل از تصویر این کار را انجام می‌دهیم:

```python
first_pixle_original_value = gray_img[0, 0]

first_pixle_quantized_value = math.floor(first_pixle_original_value / 4)

print(first_pixle_original_value, first_pixle_quantized_value)
```

به طور مشابه همین کار را برای تمام پیکسل‌ها انجام می‌دهیم:

```python
new_img = gray_img

for i in range(0, gray_img.shape[0]):
    for j in range(0, gray_img.shape[1]):
        new_img[i, j] = math.floor(gray_img[i, j] / 4)

plt.axis('off')
plt.imshow(new_img, cmap='gray')
plt.show()
```
