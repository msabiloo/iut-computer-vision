# نمونه‌برداری (Sampling)
به زبان ساده یعنی هر تصویر چه تعداد پیکسل به صورت افقی و چه تعداد پیکسل به صورت عمودی داشته باشد.

برای مثال یک تصویر را فراخوانی می‌کنیم و می‌خواهیم ببینیم تصویر چند در چند است

```python
import cv2 as cv
from matplotlib import pyplot as plt

img = cv.imread('images/TrinityRegentHouse.jpg')

plt.axis('off')
plt.imshow(cv.cvtColor(img, cv.COLOR_BGR2RGB))
plt.show()

print(img.shape)
```

با استفاده از پراپرتی shape اطلاعات زیر را از تصویر خواهیم داشت:
* تعداد پیکسل‌های عمودی تصویر
* تعداد پیکسل های افقی تصویر
* تعداد کانال‌های تصویر

```python
height = img.shape[0]

width = img.shape[1]

print(width, height)
```

معمولا وقتی می گوییم عکس چند در چند است، ابتدا تعداد پیکسل‌ها به صورت افقی را می‌گوییم و بعد تعداد پیکسل‌ها به صورت عمودی.

در OpenCV برعکس است؛ یعنی ابتدا تعداد پیکسل‌های افقی () و بعد تعداد پیکسل‌های عمودی است.

تصویر اولیه را به حالت تصویر خاکستری درمی‌آوریم:

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

plt.axis('off')
plt.imshow(gray_img, cmap='gray')
plt.show
```

برای resampling می‌توانیم از تابع resize استفاده کنیم:

```python
_256_192_gray_img = cv.resize(gray_img, (256, 192))
_128_96_gray_img = cv.resize(gray_img, (128, 96))
_64_48_gray_img = cv.resize(gray_img, (64, 48))
_32_24_gray_img = cv.resize(gray_img, (32, 24))

f, axarr = plt.subplots(2, 2)

axarr[0, 0].imshow(_256_192_gray_img, cmap='gray')
axarr[0, 0].set_title('256 × 192')
axarr[0, 0].axis('off')

axarr[0, 1].imshow(_128_96_gray_img, cmap='gray')
axarr[0, 1].set_title('128 × 96')
axarr[0, 1].axis('off')

axarr[1, 0].imshow(_64_48_gray_img, cmap='gray')
axarr[1, 0].set_title('64 × 48')
axarr[1, 0].axis('off')

axarr[1, 1].imshow(_32_24_gray_img, cmap='gray')
axarr[1, 1].set_title('32 × 24')
axarr[1, 1].axis('off')

f.subplots_adjust(hspace=0.3)
plt.show()
```
